# Módulo 2 — Flask: Cadastro, Login e Home “Olá, {nome}”

### **Objetivo**:

Criar uma págian web que contenha cadastro, login e página inicial.


Este notebook acompanha o projeto prático contido no arquivo ZIP. Siga o README do projeto para configurar venv, instalar dependências, criar `.env`, rodar migrações e executar a aplicação.

## **Como criar um projeto do absoluto 0**

### **1) Preparar ambiente**<br>
No terminal, dentro de uma pasta vazia do projeto:

### PYTHON

**Windows:**
```python

python --version
pip --version
python -m venv .venv
.venv\Scripts\Activate.ps1
pip install --upgrade pip
pip install Flask


```
**MacOS:**

```python

python3 --version
python3 -m venv .venv
source .venv/bin/activate
python -m pip install --upgrade pip
pip install Flask

```

**Linux (Debian/Ubuntu):**

```python
sudo apt update && sudo apt install -y python3 python3-venv python3-pip
python3 -m venv .venv
source .venv/bin/activate
python -m pip install --upgrade pip
pip install Flask

```



### CONDA

```python
conda create -n flaskapp python=3.11
conda init powershell
conda activate flaskapp

```

Nota: depois de ativar a venv, o prompt costuma mostrar (.venv) no começo.

### **2) Instalar dependências**<br>
No terminal, dentro de uma pasta vazia do projeto:

### PYTHON

**Python:**
```python

python -m pip install --upgrade pip
pip install Flask Flask-WTF Flask-Login Flask-SQLAlchemy email-validator


```
**Conda:**

```python

conda install -c conda-forge flask flask-wtf flask-login flask-sqlalchemy email-validator python-dotenv
# (se faltar algo no conda)
pip install flask-migrate



```

### **2) Criar a página**<br>

Dentro da pasta que contém os módulos crie a seguitne extrutura de pastas

### PASTAS

```powershell
flask_do_zero/
  app.py
  templates/
    base.html
    auth/
      login.html
      register.html
    main/
      home.html

```

### **3) Escreva o código**<br>

Escreva o código para realização da tarefa

Em app.py será escrita a aplicação Flask com rotas (register, login, home, logout), modelo User e formulários.

Em base.html será escrito o layout base com CSS simples, mensagens flash e bloco de conteúdo.

Em login.html será escrito o HTML de login contendo os campos de login (e-mail, senha, lembrar-me).

Em register.html será escrito o HTML de registro contendo os campos de registro (nome, e-mail, senha, confirmar).

Em home.html será escrito o HTML da página inicial protegida mostrando “Olá, {nome}” e botão de sair.

### PASTAS

```powershell
flask_do_zero/
  app.py
  templates/
    base.html
    auth/
      login.html
      register.html
    main/
      home.html

```
