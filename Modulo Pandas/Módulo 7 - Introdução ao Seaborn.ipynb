# Introdu√ß√£o ao Matplotlib

O **Matplotlib** √© a biblioteca de visualiza√ß√£o de dados mais tradicional e poderosa do Python. Criada por **John D. Hunter**, ela permite criar gr√°ficos 2D e 3D altamente personaliz√°veis.

### üîß Caracter√≠sticas:
- Muito flex√≠vel: permite controlar cada elemento do gr√°fico (cor, eixos, estilo, anota√ß√µes, etc.).
- Base para outras bibliotecas (Seaborn, Pandas Plot, etc.).
- Produz gr√°ficos para relat√≥rios cient√≠ficos, engenharia, estat√≠stica e mais.

No entanto, por ser t√£o flex√≠vel, pode exigir mais c√≥digo para fazer gr√°ficos simples.


# Introdu√ß√£o ao Seaborn

O **Seaborn** √© uma biblioteca de visualiza√ß√£o estat√≠stica baseada no Matplotlib. Criada por **Michael Waskom**, ela facilita a cria√ß√£o de gr√°ficos bonitos com menos c√≥digo, especialmente a partir de `DataFrames`.

### üéØ Objetivo principal:
Permitir gr√°ficos estat√≠sticos padronizados e prontos para an√°lise de dados.

### üåü Vantagens:
- Integra√ß√£o nativa com `pandas`.
- Suporte autom√°tico a estat√≠sticas (m√©dias, regress√µes, distribui√ß√£o).
- Estilos visuais bonitos por padr√£o.
- √ìtimo para gr√°ficos r√°pidos e de alto n√≠vel.

Seaborn **n√£o substitui** o Matplotlib, mas o complementa.


# Matplotlib vs Seaborn

| Aspecto                         | Matplotlib                                     | Seaborn                                           |
|----------------------------------|------------------------------------------------|----------------------------------------------------|
| N√≠vel de controle                | Muito alto (baixo n√≠vel)                      | M√©dio (alto n√≠vel, com base no matplotlib)         |
| Integra√ß√£o com pandas            | Manual (com `.plot()`)                        | Autom√°tica (`data=`, `x=`, `y=`)                   |
| Est√©tica dos gr√°ficos padr√£o     | Simples                                       | Bonita e informativa                              |
| Cria√ß√£o de gr√°ficos estat√≠sticos | Manual (ex: boxplot exige dados agrupados)   | Autom√°tica (ex: `sns.boxplot()`)                  |
| Facilidade de uso                | Exige mais linhas de c√≥digo                   | R√°pido e conciso                                   |
| Personaliza√ß√£o                   | Total, mas verbosa                            | Menor, mas com estilo moderno                     |
| Casos de uso                     | Publica√ß√µes cient√≠ficas, relat√≥rios t√©cnicos  | Explora√ß√£o e an√°lise de dados                     |

## 3. Estilos Globais e Paletas no Seaborn

O Seaborn permite configurar estilos de gr√°ficos globalmente, como:

Tamb√©m podemos definir paletas de cores personalizadas como:

Essas configura√ß√µes afetam todos os gr√°ficos seguintes no notebook.


```python
# Definindo estilo e paleta
sns.set_style("whitegrid")
sns.set_palette("pastel")

# Exemplo: Boxplot estilizado
sns.boxplot(data=df, x='Categoria', y='Valor', hue='Grupo')
plt.title("Boxplot com estilo personalizado")
plt.show()
```
